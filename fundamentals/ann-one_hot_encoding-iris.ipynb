# install libraries

```
import tensorflow as tf
print(tf.__version__)
import tensorflow_datasets as tfds
```

```
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if logs.get('acc') >= .99:
      self.model.stop_training = True

callbacks = myCallback()
```

# get dataset

```
# as_supervised set
# data = tfds.load('iris', as_supervised=True, split='train[:80%]')

# WITHOUT as_supervised set
data = tfds.load('iris', split='train[:80%]', shuffle_files=True)
```

# define helper functions 

```
# as_supervised set
# def preprocess(f, label):
#     return f, tf.one_hot(label,3)

# WITHOUT as_supervised set
def preprocess(features):
    return features['features'], tf.one_hot(features['label'], 3)
```

```
def main():
    # batch() – this allows you to consume the data from your TensorFlow Dataset in batches
    # map() – this allows you to transform the data using lambda statements applied to each element
    train_set = data.map(preprocess).batch(10)
    print(train_set)
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(units=10, activation='relu', input_shape=(4,)),
        tf.keras.layers.Dense(units=10, activation='relu'),
        tf.keras.layers.Dense(units=3, activation='softmax')
    ])
    model.summary()
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
    model.fit(train_set, epochs=100, callbacks=[callbacks])

main()
```
