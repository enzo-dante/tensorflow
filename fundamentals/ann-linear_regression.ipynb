# import libraries

```
import tensorflow as tf
import numpy as np
from tensorflow import keras 
```

# get dataset

```
# dataset
xs = [1, 2, 3, 4, 5]
ys = [100, 150, 200, 250, 300]

# preprocess into 1d numpy arrays for model processing
xs = np.array(xs)
ys = np.array(ys)
```

# define the model



```
model = keras.Sequential([
  # a fully-connected dense layer that maps inputs to outputs to optimize weights
  keras.layers.Dense(units=1, input_shape=[1])
])
model.summary()
```

# compile the model

```
# build the model by compiling it with a loss, optimizer, and objective metrics
model.compile(loss='mse', optimizer=keras.optimizers.SGD(), metrics=['acc'])
```

# train the model 

y = 50 + 50x

`ex: 150 = 50 + 50(2)`


```
# fit the model to train & learn the relationships
# assign to a history object for performance querying
# .fit(observations, labels, epochs, callbacks, validation_dataset, verbose_output)
history = model.fit(xs, ys, epochs=500)
```

# validate trained model

```
print(model.predict([7]))
```
